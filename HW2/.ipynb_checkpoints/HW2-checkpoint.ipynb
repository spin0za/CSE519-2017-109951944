```python
import pandas as pd
import numpy as np
# import scipy.stats as sps
import seaborn as sns
import matplotlib.pyplot as plt
# import matplotlib.dates as mdates
%matplotlib notebook
import random as rd
# import datetime as dt
```

```python
# read original data save to a faster format
df = pd.read_csv('properties_2016.csv', low_memory=False)
df.to_pickle('properties_2016.pkl')
```

```python
# initialize
df0 = pd.read_pickle('properties_2016.pkl')
df = df0
```

```python
# cleaning
# delete entries where room number = 0, not realistic
roomcnts = [col for col in df if 'roomcnt' in col]
df[roomcnts] = df[roomcnts].replace(0.0, np.nan)

# drop empty records(rows)
df.drop('parcelid', axis=1, inplace=True)
df.dropna(how='all', inplace=True)

# drop categorical data
# NOTE: building quality id has numeric meaning!
keyword = ['id', 'fips', 'census', 'property']
df.drop([col for col in df if any(w in col for w in keyword)], axis=1, inplace=True)

# drop columns with 90% missing
df.drop([col for col in df if df[col].count() / df.shape[0] < .1], axis=1, inplace=True)

# drop columns with almost constant value
df.drop([col for col in df if df[col].value_counts().iloc[0] / df[col].count() > .95], axis=1, inplace=True)
```

```python
df.count()
```

```python
corrmat = df.corr()
```

```python
plt.subplots()
sns.heatmap(corrmat, cmap='bwr', xticklabels=True, yticklabels=True)
plt.yticks(rotation=0);
plt.xticks(rotation=90);
plt.show()
```

```python
# line chart
plt.subplots()
x = 'yearbuilt'; y = 'buildingqualitytypeid'
df.groupby([x])[y].mean().plot()
plt.xlabel(x); plt.ylabel(y)
plt.title('Figure 2.1')
plt.show()
```

```python
# line chart
plt.subplots()
x = 'yearbuilt'; y = 'number of houses'
df.groupby([x])[x].count().plot()
plt.xlabel(x); plt.ylabel(y)
plt.show()
```
